# K-Means Clustering

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3,4]].values   
# We can give index like 3:5 or the list[3,4]
# just 2 features selected to check the plot. In realty, all features must be selected
```

```
#print(X)
```

## Using the elbow method to find the optimal number of clusters

```
#Within-Cluster-Sum-of-Squares(WCSS)
from sklearn.cluster import KMeans
wcss = []

for i in range(1,11):
  kmeans = KMeans(n_clusters=i,init= 'k-means++',random_state=42)
  # 42 brings luck in mathematics... hahaha
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('number of clusters')
plt.ylabel('Within-Cluster-Sum-of-Squares(WCSS)')
plt.show()
```

## Training the K-Means model on the dataset

```
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
# fit_predict method trains the model and returns the 5 different values for each cluster. dependent variable.
# each data point belonging to which cluster.
print(y_kmeans)
```

## Visualising the clusters

```
plt.scatter(X[y_kmeans == 0,0],X[y_kmeans == 0,1],s=25,c='red',label='Cluster 1') # s is size
plt.scatter(X[y_kmeans == 1,0],X[y_kmeans == 1,1],s=25,c='blue',label='Cluster 2')
plt.scatter(X[y_kmeans == 2,0],X[y_kmeans == 2,1],s=25,c='green',label='Cluster 2')
plt.scatter(X[y_kmeans == 3,0],X[y_kmeans == 3,1],s=25,c='cyan',label='Cluster 4')
plt.scatter(X[y_kmeans == 4,0],X[y_kmeans == 4,1],s=25,c='magenta',label='Cluster 5')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='black', s=75, label='Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
fig = plt.gcf()
fig.set_size_inches(11,8)
plt.show()
```
